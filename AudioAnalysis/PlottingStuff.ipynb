```python
import matplotlib.pyplot as plt
%matplotlib inline
import math
import numpy as np
```

```python
values = np.zeros((512))
with open("out.txt", "r") as f:
    count = 0
    for line in f:
        values[count] = int(line[:-1].split(",")[1][1:])
        count += 1
plt.plot(values)
plt.show()
```
